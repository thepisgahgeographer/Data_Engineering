```python
bill = float(input("What was the total bill?"))
tip = int(input("What tip would you like to give? 10, 12, or 15"))
people = int(input("How many people should split the bill?"))

final_tip = tip / 100 * bill + bill
final = round(final_tip / people, 2)


print(f"Everyone should pay $: {final}")
```

```python
weight = 85
height = 1.85

bmi = weight / (height **2)

if bmi < 18.5:
    print("Underwight")
elif bmi>=18.5 and bmi<25:
    print("normal weight")
else:
    print("overweight")
```

```python
print("Welcome to Python Pizza Deliveries!")

size = input("What size pizza do you want?")
pep = input("Do you want pepperoni on your pizza?")
extra_cheese = input("Do you want extra cheese")
cost = 0

if size == "S":
    cost += 15
    if extra_cheese == "Y":
        cost += 1
    if pep == "Y":
        cost +=2
    print(f"Your final bill is: {cost}")
elif size == "M":
    cost += 20
    if extra_cheese == "Y":
        cost += 1
    if pep == "Y":
        cost +=3
    print(f"Your final bill is: {cost}")
elif size == "L":
    cost += 25
    if extra_cheese == "Y":
        cost += 1
    if pep == "Y":
        cost +=3
    print(f"Your final bill is: {cost}")



```
